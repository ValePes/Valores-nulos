```python
#eliminamos columnas innecesarias 
df2= df.drop(['scrape_id', 'last_scraped','host_id','host_url', 'region_parent_name','region_parent_parent_id', 'last_searched',
              'host_location', 'host_about', 'host_acceptance_rate', 'host_thumbnail_url', 'region_parent_id',
              'host_picture_url', 'host_neighbourhood', 'host_listings_count', 'region_parent_parent_name',
              'host_total_listings_count','host_verifications', 'host_has_profile_pic', 'minimum_minimum_nights',
              'maximum_minimum_nights', 'minimum_maximum_nights','maximum_maximum_nights','minimum_nights_avg_ntm',
              'maximum_nights_avg_ntm', 'calendar_last_scraped','calendar_updated','license', 'requires_license'], axis=1)
df2
```

```python
df2.info()
```

```python
#identificamos valores nulos
val_nul=df2.isnull().sum()
val_nul
```

```python
#identificamos valores nulos del dataframe 
val_nul=df2.isnull().sum().sum()
val_nul
```

```python
#hacemos una copia de df2
df3=df2.copy()
```

```python
#reviews_per_month. Para esta variable, debido a los datos que maneja es conveniente utilizar promedio para añadir valores más claros de acuerdo a las 
#tendencias en reseñas que el cliente puede encontrar
df3["reviews_per_month"]=df3["reviews_per_month"].fillna(round(df2["reviews_per_month"].mean(),1))
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```

```python
#para las variables review_scores_rating, review_scores_accuracy , review_scores_location, review_scores_value,es importante aplicar una sustitución 
#por promedio, debido a que sus datos son necesarios para un análisis adecuado de las necesidades del cliente, necesitamos obtener valores 
#que sean claros

df3["review_scores_rating"]=df3["review_scores_rating"].fillna(round(df2["review_scores_rating"].mean(),1))
df3["review_scores_accuracy"]=df3["review_scores_accuracy"].fillna(round(df2["review_scores_accuracy"].mean(),1))
df3["review_scores_location"]=df3["review_scores_location"].fillna(round(df2["review_scores_location"].mean(),1))
df3["review_scores_value"]=df3["review_scores_value"].fillna(round(df2["review_scores_value"].mean(),1))
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```

```python
#price.Para esta variable, debido a los datos que maneja es conveniente utilizar promedio para añadir valores más claros de acuerdo a las 
#tendencias en precios que el cliente desea encontrar
df3['price'] = pd.to_numeric(df['price'].astype(str).str.replace('$', '').str.replace(',', ''), errors='coerce')
df3['price'].fillna(round(df3['price'].mean(), 1), inplace=True)
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```

```python
#Aplicaremos metodo de sustitucion por mediana para los sisguiente4s valores
#bathrooms. Para esta variable aplicamos método de sustitución por mediana debido a que presenta valores discretos, por ende su sustitución será robusta 
df3["bathrooms"]=df3["bathrooms"].fillna(round(df2["bathrooms"].median(), 1))
#bedrooms. Para esta variable aplicamos método de sustitución por mediana debido a que presenta valores discretos, por ende su sustitución será robusta ya que no se manejan datos tan dispersos. 
df3["bedrooms"]=df3["bedrooms"].fillna(round(df2["bedrooms"].median(), 1))
#beds. Al igual que las variables anteriores, al ser valores numéricos pequeños utilizar la mediana es de gran utilidad y obtendríamos valores típicos
df3["beds"]=df3["beds"].fillna(round(df2["beds"].median(), 1))
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```

```python
#Aplicaremos metodo de sustitucion por valor numerico concreto para las variables review_scores_cleanliness, review_scores_checkin y review_scores_communication utilizamos 
#valores numéricos exactos dentro del rango en que se encuentran sus datos con el fin de que los valores a modificar no presenten un desbalance entre los ya existentes
df3["review_scores_cleanliness"]=df3["review_scores_cleanliness"].fillna(5)
df3["review_scores_checkin"]=df3["review_scores_checkin"].fillna(5)
df3["review_scores_communication"]=df3["review_scores_communication"].fillna(5)
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```

**Sustitucion por un string en concreto**

```python
#Para la variable descripción, al ser un objeto donde se almacenan cadenas de caracteres mismas que podrían afectar los datos proporcionados es recomendable mantener su uniformidad sustituyendo el valor por un string en concreto descriptivo de la ausencia del dato. 
df3["description"] = df3["description"].fillna("No se proporciona descripción") 
#Para neighborhood_overview utilizamos sustitución por string debido a que los datos presentados no pueden ser alterados para una descripción importante indicaremos por ello que no se presenta la información
df3["neighborhood_overview"] = df3["neighborhood_overview"].fillna("No se proporcionan detalles")
#En neighbourhood, podemos sustituir or un string en concreto, debido a que los datos contenidos en esa columna se repiten sustituir por un string entonces no afectaría los datos presentados si lo hacemos por la ciudad donde se encuentra el hospedaje. 
df3["neighbourhood"] = df3["neighbourhood"].fillna("Mexico City, Mexico")
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```

```python
#host_response_time al ser una variable tipo objeto que almacena cadenas de caracteres el método de sustitución por un string en concreto  resuta util para indicar que el tiempo de respuesta del host es no fue indicado y no alterar los datos. 
df3["host_response_time"] = df3["host_response_time"].fillna("Desconocido")
#host_response_rate. Al igual que la variable anterior, utilizamos um string para indicar que el tiempo de respuesta del anfitrión no fue indicado por el huesped con un -% y de esta manera al obtener resultados la tasa del porcentaje no se vea afectada
df3["host_response_rate"] = df3["host_response_rate"].fillna("-%")
#first_review. Para esta variable, como se trabaja con fechas utilizare un string que indique la primera revisión con la fecha de cuando se están llenando los datos faltantes, para saber a que datos hay que ponerles mayor atención. 
df3["first_review"] = df3["first_review"].fillna("23/01/2025")
#last_review. También es una variable que trabaja con fechas, por ello se utilizará un string en concreto para definir una fecha como ultima revisión y saber que filas han sido modificadas 
df3["last_review"] = df3["last_review"].fillna("24/01/2025")
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```

**sUSTITUCIÓN DE VALORES NULOS POR VALORES NO NULOS HACIA DELANTE O HACIA ATRÁs**

```python
#para la variable host_is_superhost utilizamos bfill debido a que los datos que acepta esta variable son unicamente dos y estos pueden ser remplazados por los valores posteriores
df3["host_is_superhost"] = df3["host_is_superhost"].fillna(method="bfill")
#para la variable has_availability, al igual que la anterior, unicamente acepta dos valores, verdadero o falso, por lo tanto puede ser reemplazada por los valores anteriores a ella sin afectar los datos. 
df3["has_availability"] = df3["has_availability"].fillna(method="bfill")
#para la variable bathrooms_text podemos aplicar metodo de sustitución hacia atrás debido a que los datos que trabaja se encuentran dentro del mismo rango.  
df3["bathrooms_text"] = df3["bathrooms_text"].fillna(method="bfill")
```

```python
#corroboramos valores nulos 
val_nul=df3.isnull().sum()
val_nul
```
